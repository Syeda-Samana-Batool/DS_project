# SHEET NO 1 : TimeSeries_KeyIndicators

```python
sheet1= pd.read_excel('D:\Covid-19\COVID_FINAL_DATA.xlsx',sheet_name='TimeSeries_KeyIndicators')
```

```python
sheet1.head(10)
```

```python
sheet1.isnull().sum()  #checking Null values
```

```python
sheet1.columns=sheet1.columns.str.strip().str.replace(' ','_')  #replacing spaces with hypens
```

```python
sheet1.head(10)
```

```python
sheet1.drop('Unnamed:_0',axis=1,inplace=True) #dropping first column
```

```python
sheet1.head()
```

```python
sheet1['Home_Quarantine'].fillna(0,inplace=True)  #filling empty spaces of Home_quarantine with zeros
sheet1['Home_Quarantine']
```

```python
sheet1.isnull().sum() #zero null values 
```

```python
sheet1['Home_Quarantine']=sheet1['Home_Quarantine'].apply(np.int64)
```

```python
sheet1['Date']=pd.to_datetime(sheet1.Date) #setting date format
```

```python
sheet1.dtypes
```

```python
sheet1['Tests__performed_in_last_24_hours']=np.where(sheet1['Tests__performed_in_last_24_hours']=='No 0',0,sheet1.Tests__performed_in_last_24_hours)
```

```python
sheet1['Tests__performed_in_last_24_hours']=sheet1['Tests__performed_in_last_24_hours'].apply(np.int64)  
```

```python
sheet1.dtypes
```

```python
sheet1.duplicated().any() #checking duplicate columns
```

```python
sheet1.describe() #sheet 1 is cleaned
```

```python
sheet1.to_excel('D:\TimeSeries_KeyIndicators.xlsx') #saving back the file as an excel sheet
```
